```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# Test the sigmoid function with different values of x
print(sigmoid(0))    # Output: 0.5
print(sigmoid(2))    # Output: 0.8807970779778823
print(sigmoid(-2))   # Output: 0.11920292202211755
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate an array of x values
x_values = np.linspace(-10, 10, 100)

# Apply the sigmoid function to each x value
y_values = [sigmoid(x) for x in x_values]

# Plot the results
plt.plot(x_values, y_values)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.show()
```

# Step 2: Generate a Synthetic Dataset

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """
    Compute the sigmoid of x.
    """
    return 1 / (1 + np.exp(-x))

# Generate a range of values from -10 to 10
x_values = np.linspace(-10, 10, 100)
# Compute the sigmoid for each value
y_values = sigmoid(x_values)

# Plotting the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Generate a synthetic dataset for binary classification
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, random_state=42)

# Fit a logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Get the decision boundary for plotting
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Create a grid of points to evaluate the model
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Compute the model's decision function
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the data points and decision boundary
plt.figure(figsize=(10, 6))

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=50)

# Plot decision boundary
plt.contourf(xx, yy, sigmoid(Z), alpha=0.3, cmap='bwr', levels=[0, 0.5, 1])
plt.contour(xx, yy, Z, levels=[0], colors='k', linestyles='--')

plt.title('Sigmoid Function in Logistic Regression')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Probability')
plt.grid(True)
plt.show()

# Print model accuracy
y_pred = model.predict(X)
print(f"Model Accuracy: {accuracy_score(y, y_pred):.2f}")
```
